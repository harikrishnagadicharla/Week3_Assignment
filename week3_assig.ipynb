```python
from gensim.models import Word2Vec

# Define the sentences
sentences = [
    "Artificial intelligence is the future of technology",
    "Machine learning helps computers learn patterns",
    "Deep learning uses neural networks for decision-making",
    "Natural language processing improves communication between humans and machines",
    "Word embeddings capture the meaning of words in a vector space"
]

# Tokenize the sentences
tokenized_sentences = [sentence.lower().split() for sentence in sentences]

# Train a Word2Vec model
model = Word2Vec(sentences=tokenized_sentences, vector_size=10, window=3, min_count=1, workers=1)

# Extract the embedding for the word "machine"
embedding_machine = model.wv["machine"]
print("Embedding for 'machine':", embedding_machine)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
input_shape = (10, 20)
n_classes = 5
model = Sequential([
    SimpleRNN(64, activation='tanh', return_sequences=True, input_shape=input_shape),
    SimpleRNN(32, activation='tanh', return_sequences=True),
    SimpleRNN(16, activation='tanh'),
    Dense(n_classes, activation='softmax')])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the T5-XL model and tokenizer
model_name = "google/t5-xl-lm-adapt"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def generate_response(prompt, max_length=100):
    """Generates a response using the T5 model."""
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids, max_length=max_length)
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

# One-shot prompting
one_shot_prompt = "Medical Diagnosis: Patient has fever, sore throat, and fatigue. What is the possible condition?"
response_one_shot = generate_response(one_shot_prompt)
print("One-Shot Response:", response_one_shot)

# Few-shot prompting with examples
few_shot_prompt = (
    "Medical Diagnosis: Patient has fever and cough. Possible condition: Flu.\n"
    "Medical Diagnosis: Patient has headache and dizziness. Possible condition: Migraine.\n"
    "Medical Diagnosis: Patient has chest pain and shortness of breath. Possible condition: Heart attack.\n"
    "Medical Diagnosis: Patient has fever, sore throat, and fatigue. Possible condition:"
)

response_few_shot = generate_response(few_shot_prompt)
print("Few-Shot Response:", response_few_shot)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load an open-source CoT reasoning model for math
model_name = "Qwen/Qwen2.5-Math-1.5B"  # Change model if needed
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

def generate_response(prompt, max_length=200):
    """Generates a response using the loaded model."""
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids, max_length=max_length)
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Chain of Thought (CoT) prompting for a math problem
cot_prompt = (
    "Solve step by step: If a train travels at 60 km/h for 2 hours, how far does it go?\n"
    "Let's think step by step."
)

response_cot = generate_response(cot_prompt)
print("CoT Response:", response_cot)
```
